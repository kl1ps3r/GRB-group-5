```python
import numpy as np

import matplotlib.pyplot as plt

from parameters import *
```

```python
sigma = 3

d_m = np.linspace(1.0, 1.5, 100)
d_d = 1.
D = np.linspace(0.5, 2., 100)
l_m = np.linspace(0.001, 0.01, 100)
l_d = np.linspace(0.001, 0.01, 100)
```

```python
D_index = 50

grid = np.meshgrid(d_m, d_d, D)
FCFOV = Fully_Coded(*grid)
FCFOV.shape
```

```python
plt.plot(D, FCFOV[0, 99, :])


plt.title(r"Mask detector seperation (D) against FCFOV for $d_d = 1.0$m and $d_m = 1.5$m")
plt.xlabel("D [m]")
plt.ylabel(r"FCFOV [deg]")

plt.savefig("./mask_parameter_plots/FCFOV.png")
```

```python
grid2 = np.meshgrid(d_m, d_d, D)
PCFOV = Partially_Coded(*grid2)
```

```python
plt.plot(D, PCFOV[0, 99, :])

plt.title(r"Mask detector seperation (D) against PCFOV for $d_d = 1.0$m and $d_m = 1.5$m")
plt.xlabel("D [m]")
plt.ylabel(r"PCFOV [deg]")

plt.savefig("./mask_parameter_plots/PCFOV.png")
```

```python
grid3 = np.meshgrid(l_m, D)

ang_res = AngRes(*grid3) * 60
```

```python
plt.pcolormesh(l_m, D, ang_res[:-1, :-1], cmap="RdBu", shading="flat")
plt.colorbar()

plt.title("Angular Resolution Colourmap [arcmin]")
plt.xlabel(r"$l_m$ [m]")
plt.ylabel("$D$ [m]")

plt.savefig("./mask_parameter_plots/AngRes.png")
```
