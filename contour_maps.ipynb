```python
import numpy as np

import matplotlib.pyplot as plt

from parameters import *
```

```python
sigma = 3

d_m = np.linspace(0.5, 2.0, 100)
d_d = 1.
D = np.linspace(0.5, 2., 100)
l_m = np.linspace(0.001, 0.01, 100)
l_d = np.linspace(0.001, 0.01, 100)
```

```python
D_index = 50

grid = np.meshgrid(d_m, d_d, D)
FCFOV = np.rad2deg(Fully_Coded(*grid))
d_m[99]
```

```python
plt.plot(D, FCFOV[0, 99, :])


plt.title(f"Mask detector seperation (D) against FCFOV for $d_m = {d_m[-1]}$m and $d_d = {d_d}$m")
plt.xlabel("D [m]")
plt.ylabel(r"FCFOV [deg]")

plt.savefig("./mask_parameter_plots/FCFOV.png")
```

```python
grid2 = np.meshgrid(d_m, d_d, D)
PCFOV = np.rad2deg(Partially_Coded(*grid2))
```

```python
plt.plot(D, PCFOV[0, 99, :])

plt.title(f"Mask detector seperation (D) against PCFOV for $d_m = {d_m[-1]}$m and $d_d = {d_d}$m")
plt.xlabel("D [m]")
plt.ylabel(r"PCFOV [deg]")

plt.savefig("./mask_parameter_plots/PCFOV.png")
```

```python
grid3 = np.meshgrid(l_m, D)

ang_res = np.rad2deg(AngRes(0.01, 1)) * 60
PSL = np.rad2deg(Localisation(0.01, 1, 3)) * 60
```
